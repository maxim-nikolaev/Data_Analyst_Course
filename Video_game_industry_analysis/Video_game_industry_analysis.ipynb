#  <a name="5">3. Составление портрета потребителя</a>

### Топ-5 игровых платформ по регионам

В наборе данных представлены объемы продаж в следующих регионах: США, Европа, Япония и другие страны.

Определим топ-5 платформ для каждого из регионов.



США

```python
USA_top5 = df_act.pivot_table(index='platform', values='na_sales', aggfunc=['sum']).reset_index() # сводная таблица 
USA_top5.columns = ['platform','total_na_sales'] # определяем столбцы для сводной таблицы
USA_top5['market_share'] = USA_top5['total_na_sales']/(USA_top5['total_na_sales']).sum() # рассчитваем составляющую продаж 
USA_top5 = USA_top5.sort_values(by='total_na_sales', ascending=False).reset_index(drop=True)

others_sales = USA_top5.loc[5:,'total_na_sales'].sum() 
others_share = USA_top5.loc[5:,'market_share'].sum() 

USA_top5 = USA_top5[0:5].append({'platform':'others', 'total_na_sales':others_sales, 'market_share':others_share},
                                ignore_index=True)
```

Европа

```python
EU_top5 = df_act.pivot_table(index='platform', values='eu_sales', aggfunc=['sum']).reset_index()
EU_top5.columns = ['platform','total_eu_sales']
EU_top5['market_share'] = EU_top5['total_eu_sales']/(EU_top5['total_eu_sales']).sum()
EU_top5 = EU_top5.sort_values(by='total_eu_sales', ascending=False).reset_index(drop=True)

others_sales = EU_top5.loc[5:,'total_eu_sales'].sum()
others_share = EU_top5.loc[5:,'market_share'].sum()

EU_top5 = EU_top5[0:5].append({'platform':'others', 'total_eu_sales':others_sales, 'market_share':others_share},
                              ignore_index=True)
```

Япония

## <a name="7">5. Проверка гипотез</a>

### Продажи приставок Wii и WiiU одинаковы

**Выдвинем нулевую и альтернативную гипотезы**

**H0**: средние значения объема продаж в мире приставок «Wii» и «WiiU» одинаковы.

**H1**: средние значения объема продаж в мире приставок «Wii» и «WiiU» не одинаковы (мы не оцениваем у какой больше).

Устанавливаем критерий значимости **alpha** для решения: уровень значимости обычно устанавливается в 5%. Также выберем этот уровень. Когда вероятность получения выборочного среднего меньше 5%, если нулевая гипотеза верна, мы отвергаем значение, указанное в нулевой гипотезе.

Чтобы проверить нашу гипотезу о том, что средние значения двух статистических совокупностей равны на основе взятых из них выборок, применим метод **scipy.stats.ttest_ind()**.

Создадим две выборки: мировой объем продаж каждой из игровых платформ (выборки делаем в период с 2012 по 2016 годы - набор данных df_act).

```python
sample_1 = df[df['platform'] == 'Wii'].dropna()['global_sales']
sample_2 = df[df['platform'] == 'WiiU'].dropna()['global_sales']
```

Выполним тест Левена на оценку равности дисперсий наших выборок.

```python
st.levene(sample_1, sample_2)
```

P-значение критерия Левена больше 5% означает, что у выборок дисперсии не равны. Следовательно, мы установим для параметра equal_var значение False.

Рассчитаем T-критерий для средних значений двух выборок.

```python
alpha = .05

results = st.ttest_ind(
                        sample_1, 
                        sample_2, 
                        equal_var=False)

print('p-value: ', results.pvalue)

if (results.pvalue < alpha):
        print("Нулевая гипотеза опровергнута")
else:
        print("Нулевая гипотеза подтверждена")
```

По результатам тестовой статистики методом t-test подтверждаем нулевую гипотезу, т.к. pvalue больше выбранного уровня значимости alpha. Значит, общемировые продажи игровых платформ «Wii» и «WiiU» одинаковы.

2. По объему продаж во всем мире лидерами являются следующие 5 игровых платформ: «PS4», «PS3», «X360», «3DS», «XOne».
3. В среднем на появление новых платформ и исчезновение старых уходит около 10 лет, поэтому для более актуального анализа взяты данные за период с 2012 года.
4. Существенных различий в продажах для топ-5 платформ нет. 
5. Существенных различий между общим объемом продаж одних и тех же игр на каждой платформе нет.
6. Игры в жанре экшн самые прибыльные и востребованные. 
7. Более активные игры (боевики, шутеры, спорт) имеют высокие продажи, а более интеллектуальные игры (головоломки, стратегии) — низкие.

**В третьем разделе** были проанализированы 5 самых популярных платформ, жанров и компаний, выпускающих видеоигры, в каждом регионе. В наборе данных представлены следующие регионы: США, Европа и Япония.  
1. В каждом регионе предпочтение отдается разным игровым платформам — в США большинство игр продается на платформе «X360», в Европе - на платформе «PS4», в Японии - на платформе «3DS». Интересно, что самая популярная игровая платформа в США практически не представлена в Японии. На самые популярные платформы приходится четверть всех продаж в Северной Америке и Европе и почти половина в Японии.
2. Игры в жанре экшн самые популярные как в Северной Америке, так и в Европе, в то время как в Японии предпочтение отдано ролевым играм.
3. В США самые популярные игры - выпущенные компанией Activision, в Европе - компанией Electronic Arts. Япония снова отличилась:  2 лидирующие компании в США и Европе вовсе не представлены в топе предпочтений японцев - на первом месте компания Nintendo.
4. Обобщая этот раздел, типичный пользователь в каждом регионе имеет следующие характеристики:  
    США: игрок, предпочитающий в основном в игры в жанре экшн на платформе «X360».  
    Европа: игрок, предпочитающий в основном в игры в жанре экшн на платформе «PS4».  
    Япония: игрок, предпочитающий в основном ролевые игры на платформе «3DS».  

**В четвертом разделе** были проанализированы коррелирующие величины с помощью тепловой карты. Для предсказания величины общемировых продаж были выбраны продажи в США и Европе (значение корелляционного коэффициента было выше 0,9) была построена множественная линейная регрессия, в том числе была выполнена визуализация 3D с интерактивным взаимодействием с помощью  библиотеки plotly.

**В пятом разделе** была выполнена проверка статистических гипотез.  
Мы проверили две гипотезы:
1. Продажи приставок Wii и WiiU одинаковы: данная гипотеза была выдвинута, потому что Wii была очень популярна, а WiiU является ее следующим поколением - соответственно должна быть так же востребована у игроков. Нулевая гипотеза была подтверждена, соответственно данное утверждение верно.
2. Продажи компаний Nintendo и Activision одинаковы: данная гипотеза была выдвинута потому что игры этих компаний очень популярны во всех регионах. Нулевая гипотеза была опровергнута , соответственно Продажи компаний Nintendo и Activision не одинаковы.
