На самом деле мы можем работать с этими значениями и вручную настраивать тип данных для каждого столбца в каждом конкретном случае. 

Чтобы облегчить наш дальнейший анализ, лучше всего изменить значения на логические значения True и False, где, если в строке присутствует «нет», оно будет принято как «Ложь», чтобы показать «отсутствие» элемента и ' Правда' в противном случае. Для их переноса мы будем использовать приведенные ниже функции в нашей пользовательской функции _'change_toboolean':

pandas.DataFrame.apply 

Формат: DataFrame.apply(func, axis=0, raw=False, result_type=None, args=(), **kwds) 

Применем функцию вдоль оси DataFrame.

и лямбда, представляющая анонимную функцию, где, если она используется с предыдущим pd.Series.apply, каждый элемент серии передается в эту лямбда-функцию. Здесь мы используем это для нашего условия if-else, упомянутого выше. Результатом будет еще один pd.Series, в котором каждый элемент проходит через эту лямбду.

Чтобы проверить вывод позже, а также проверить исходные данные, мы будем использовать pandas.Series.value_counts

Формат: Series.value_counts(normalize=False, sort=True, по возрастанию=False, bins=None, dropna=True) Возвращает серию, содержащую количество уникальных значений. Результирующий объект будет располагаться в порядке убывания, поэтому первый элемент является наиболее часто встречающимся элементом. Исключает значения NA по умолчанию.

```python
def change_to_boolean(col_series):
    return col_series.apply(lambda x: False if 'not' in x else True)
```

```python
# старые значения (для проверки)_значения: Какао-масло
chocolate_df.cocoa_butter.value_counts()
```

```python
chocolate_df['cocoa_butter'] = change_to_boolean(chocolate_df['cocoa_butter'])
chocolate_df.cocoa_butter.value_counts()
```

```python
# старые значения (для проверки)_значения: Ваниль
chocolate_df.vanilla.value_counts()
```

```python
chocolate_df['vanilla'] = change_to_boolean(chocolate_df['vanilla'])
chocolate_df.vanilla.value_counts()
```

```python
# старые значения (для проверки)_значения: Лецитин
chocolate_df.lecithin.value_counts()
```

```python
chocolate_df['lecithin'] = change_to_boolean(chocolate_df['lecithin'])
chocolate_df['lecithin'].value_counts()
```

```python
# старые значения (для проверки)_значения: Соль
chocolate_df.salt.value_counts()
```

```python
chocolate_df['salt'] = change_to_boolean(chocolate_df['salt'])
chocolate_df.salt.value_counts()
```

```python
# старые значения (для проверки)_значения: Сахар
chocolate_df.sugar.value_counts()
```

### Корреляция между разными столбцами 

Чтобы увидеть взаимозависимость между двумя или более переменными, используйте функцию корреляции pandas.DataFrame.corr. Тогда мы сможем проверить все корреляции одновременно.
Формат: DataFrame.corr(method='pearson', min_periods=1) Вычисляет попарную корреляцию столбцов, за исключением значений NA/null. Возвращает: DataFrame (матрица корреляции). Метод Пирсона используется по умолчанию, но Pandas позволяет использовать другие индексы.

Положительное или отрицательное значение от 0,9 до 1 указывает на очень сильную корреляцию.

Положительный или отрицательный результат от 0,7 до 0,9 указывает на сильную корреляцию.

Положительный или отрицательный результат от 0,5 до 0,7 указывает на умеренную корреляцию. 

Положительный или отрицательный результат от 0,3 до 0,5 указывает на слабую корреляцию. 

От 0 до 0,3 положительный или отрицательный указывает на незначительную корр. 

Чтобы облегчить эту визуализацию корреляций, можно использовать цвета.

Давайте воспользуемся функцией тепловой карты в Seaborn.

тепловая карта Формат: seaborn.heatmap(data, , vmin=None, vmax=None, cmap=None, center=None, robust=False, annot=None, fmt='.2g', annot_kws=None, linewidths=0, linecolor ='white', cbar=True, cbar_kws=None, cbar_ax=None, Square=False, xticklabels='auto', yticklabels='auto', mask=None, ax=None, *kwargs) Отображает прямоугольные данные в виде матрицы с цветовой кодировкой. Аннотирует каждую ячейку числовым значением, используя целочисленное форматирование. Цветовая градация наблюдается в отношении положительных и отрицательных корреляций.

```python
fig, axis = plt.subplots(1,2,figsize=(15,5))
vendors= sns.barplot(x= countries_with_most_amount_of_vendors.index, y= "company_location",
            data= countries_with_most_amount_of_vendors, 
            palette= sns.cubehelix_palette(15, start=.5, rot=-.75, reverse= True), ax=axis[0])
producers= sns.barplot(x= highest_producers.index, y= "country_of_bean_origin",
            data= highest_producers, 
            palette= sns.cubehelix_palette(15, start=.5, rot=-.75, reverse= True), ax= axis[1])

vendors.set_title("Number of Vendors per Country")
vendors.set_xlabel("Countries")
vendors.set_ylabel("# of Vendors")
plt.setp(vendors.get_xticklabels(), rotation=90, ha='right')

producers.set_title("Number of Producers per Country")
producers.set_xlabel("Countries")
producers.set_ylabel("# of Producers")
plt.setp(producers.get_xticklabels(), rotation=90, ha='right')

plt.show()
```
