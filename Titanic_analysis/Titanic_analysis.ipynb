## 2.4. Окончательные корректировки данных (обучение и тестирование)

Основываясь на оценке отсутствующих значений в наборе данных, внесем в данные следующие изменения:
* Если "Age" отсутствует для данной строки, будем приписывать 28 (средний возраст).
* Если в строке с разделением отсутствует "Embarked", будем приписывать "S" (наиболее распространенный порт посадки).
* Будем игнорировать "Cabin" как переменную. Существует слишком много недостающих значений для вменения. Исходя из имеющейся информации, представляется, что это значение связано с классом пассажира и оплаченным тарифом.

```python
train_data = train_df.copy()
train_data["Age"].fillna(train_df["Age"].median(skipna=True), inplace=True)
train_data["Embarked"].fillna(train_df['Embarked'].value_counts().idxmax(), inplace=True)
train_data.drop('Cabin', axis=1, inplace=True)
```

## 2.4.1. Дополнительные переменные


Согласно словарю данных Kaggle, и SibSp, и Parch относятся к путешествиям с семьей. Для простоты (и для учета возможной мультиколлинеарности) я объединю влияние этих переменных в один категориальный признак: путешествовал ли этот человек один или нет.

```python
## Создадим переменную для путешествий в одиночку
train_data['TravelAlone']=np.where((train_data["SibSp"]+train_data["Parch"])>0, 0, 1)
train_data.drop('SibSp', axis=1, inplace=True)
train_data.drop('Parch', axis=1, inplace=True)
```

Cоздадим категориальные переменные для класса пассажира ("P class"), пола ("Sex") и порта посадки ("Embedded").

```python
#создадим категориальные переменные и удалим некоторые переменные
training=pd.get_dummies(train_data, columns=["Pclass","Embarked","Sex"])
training.drop('Sex_female', axis=1, inplace=True)
training.drop('PassengerId', axis=1, inplace=True)
training.drop('Name', axis=1, inplace=True)
training.drop('Ticket', axis=1, inplace=True)

final_train = training
final_train.head()
```

### Теперь применим те же изменения к тестовым данным. <br>
Применим те же изменения для "Возраста" в тестовых данных, что и для тренировочных данных (если отсутствует, возраст = 28).<br> Так же удалим переменную "Cabin" из тестовых данных, так как я решила не включать ее в анализ. <br> 
В переменной порта "Отправлено" не было пропущенных значений. <br> 
Я добавлю фиктивные переменные, чтобы завершить набор тестов. <br> 
Наконец, добавим 1 недостающее значение для "Fare" с медианой, 14.45.

```python
test_df.isnull().sum()
```

```python
test_data = test_df.copy()
test_data["Age"].fillna(train_df["Age"].median(skipna=True), inplace=True)
test_data["Fare"].fillna(train_df["Fare"].median(skipna=True), inplace=True)
test_data.drop('Cabin', axis=1, inplace=True)

test_data['TravelAlone']=np.where((test_data["SibSp"]+test_data["Parch"])>0, 0, 1)

test_data.drop('SibSp', axis=1, inplace=True)
test_data.drop('Parch', axis=1, inplace=True)

testing = pd.get_dummies(test_data, columns=["Pclass","Embarked","Sex"])
testing.drop('Sex_female', axis=1, inplace=True)
testing.drop('PassengerId', axis=1, inplace=True)
testing.drop('Name', axis=1, inplace=True)
testing.drop('Ticket', axis=1, inplace=True)

final_test = testing
final_test.head()
```

## 3.1. Исследование возраста

```python
plt.figure(figsize=(15,8))
ax = sns.kdeplot(final_train["Age"][final_train.Survived == 1], color="darkturquoise", shade=True)
sns.kdeplot(final_train["Age"][final_train.Survived == 0], color="lightcoral", shade=True)
plt.legend(['Выжившие', 'Погибшие'])
plt.title('График величины возраста выживших и погибших')
ax.set(xlabel='Age')
plt.xlim(-10,85)
plt.show()
```

Распределение по возрасту выживших и умерших на самом деле очень похоже. Одно из заметных отличий заключается в том, что среди выживших большую долю составляли дети. Пассажиры, очевидно, предприняли попытку спасти детей, предоставив им место в спасательных шлюпках.

## 4.2. Обзор процедур оценки модели

Мотивация: Нужен способ выбора между моделями машинного обучения
* Цель состоит в том, чтобы оценить вероятную производительность модели на данных вне выборки.

Первоначальная идея: обучать и тестировать на одних и тех же данных
* Но максимизация точности обучения вознаграждает слишком сложные модели, которые чрезмерно соответствуют данным обучения

Альтернативная идея: разделение обучения и тестирования
* Разделим набор данных на две части, чтобы модель можно было обучать и тестировать на разных данных.
* Точность тестирования является лучшей оценкой, чем точность обучения производительности вне выборки.
* Проблема с разделением train/test
* Это обеспечивает высокую оценку дисперсии, поскольку изменение того, какие наблюдения оказались в тестовом наборе, может значительно изменить точность тестирования.
* Точность тестирования может сильно измениться в зависимости от того, какое наблюдение оказалось в тестовом наборе

### 4.2.1. Оценка модели на основе простого разделения train/test с использованием функции `train_test_split()`

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score 
from sklearn.metrics import confusion_matrix, precision_recall_curve, roc_curve, auc, log_loss

#создадим  X (Выбранные функции) и (Выжившие)
X = final_train[Selected_features]
y = final_train['Survived']

# используйте разделение train/test с разными значениями random_state
# мы можем изменить значения random_state, которые изменяют оценки точности
# оценки сильно меняются, вот почему результаты тестирования - это оценка с высокой дисперсией.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

# проверка классификационных оценок логистической регрессии
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
y_pred_proba = logreg.predict_proba(X_test)[:, 1]
[fpr, tpr, thr] = roc_curve(y_test, y_pred_proba)
print('Train/Test результаты:')
print(logreg.__class__.__name__+" accuracy is %2.3f" % accuracy_score(y_test, y_pred))
print(logreg.__class__.__name__+" log_loss is %2.3f" % log_loss(y_test, y_pred_proba))
print(logreg.__class__.__name__+" auc is %2.3f" % auc(fpr, tpr))

idx = np.min(np.where(tpr > 0.95)) # индекс первого порога, для которого чувствительность > 0,95

plt.figure()
plt.plot(fpr, tpr, color='coral', label='ROC curve (area = %0.3f)' % auc(fpr, tpr))
plt.plot([0, 1], [0, 1], 'k--')
plt.plot([0,fpr[idx]], [tpr[idx],tpr[idx]], 'k--', color='blue')
plt.plot([fpr[idx],fpr[idx]], [0,tpr[idx]], 'k--', color='blue')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1 - specificity)', fontsize=14)
plt.ylabel('True Positive Rate (recall)', fontsize=14)
plt.title('Receiver operating characteristic (ROC) curve')
plt.legend(loc="lower right")
plt.show()

print("Использование порога %.3f " % thr[idx] + "гарантирует чувствительность %.3f " % tpr[idx] +  
      "и специфичность of %.3f" % (1-fpr[idx]) + 
      ", т.е. частота ложноположительных результатов %.2f%%." % (np.array(fpr[idx])*100))
```

### 4.2.2. Оценка модели на основе K-кратной перекрестной проверки с использованием функции`cross_val_score()`

```python
# 10-кратная логистическая регрессия с перекрестной проверкой
logreg = LogisticRegression()
# Используем функцию cross_val_score
# Мы передаем все значения X и y целиком, а не X_train или y_train, это заботится о разделении данных
# cv=10 для 10 проверок
# оценка = {'accuracy', 'neg_log_loss', 'roc_auc'} для метрики оценки - хотя их много
scores_accuracy = cross_val_score(logreg, X, y, cv=10, scoring='accuracy')
scores_log_loss = cross_val_score(logreg, X, y, cv=10, scoring='neg_log_loss')
scores_auc = cross_val_score(logreg, X, y, cv=10, scoring='roc_auc')
print('Результаты K-кратной перекрестной проверки:')
print(logreg.__class__.__name__+" средний accuracy is %2.3f" % scores_accuracy.mean())
print(logreg.__class__.__name__+" средний log_loss is %2.3f" % -scores_log_loss.mean())
print(logreg.__class__.__name__+" средний auc is %2.3f" % scores_auc.mean())
```

## 4.3. GridSearchCV оценка с использованием нескольких показателей одновременно

```python
from sklearn.model_selection import GridSearchCV

X = final_train[Selected_features]

param_grid = {'C': np.arange(1e-05, 3, 0.1)}
scoring = {'Accuracy': 'accuracy', 'AUC': 'roc_auc', 'Log_loss': 'neg_log_loss'}

gs = GridSearchCV(LogisticRegression(), return_train_score=True,
                  param_grid=param_grid, scoring=scoring, cv=10, refit='Accuracy')

gs.fit(X, y)
results = gs.cv_results_

print('='*20)
print("лучшие параметры: " + str(gs.best_estimator_))
print("лучшие параметры: " + str(gs.best_params_))
print('лучший результат:', gs.best_score_)
print('='*20)

plt.figure(figsize=(10, 10))
plt.title("GridSearchCV evaluating using multiple scorers simultaneously",fontsize=16)

plt.xlabel("Inverse of regularization strength: C")
plt.ylabel("Score")
plt.grid()

ax = plt.axes()
ax.set_xlim(0, param_grid['C'].max()) 
ax.set_ylim(0.35, 0.95)

# Получим обычный массив numpy из MaskedArray
X_axis = np.array(results['param_C'].data, dtype=float)

for scorer, color in zip(list(scoring.keys()), ['g', 'k', 'b']): 
    for sample, style in (('train', '--'), ('test', '-')):
        sample_score_mean = -results['mean_%s_%s' % (sample, scorer)] if scoring[scorer]=='neg_log_loss' else results['mean_%s_%s' % (sample, scorer)]
        sample_score_std = results['std_%s_%s' % (sample, scorer)]
        ax.fill_between(X_axis, sample_score_mean - sample_score_std,
                        sample_score_mean + sample_score_std,
                        alpha=0.1 if sample == 'test' else 0, color=color)
        ax.plot(X_axis, sample_score_mean, style, color=color,
                alpha=1 if sample == 'test' else 0.7,
                label="%s (%s)" % (scorer, sample))

    best_index = np.nonzero(results['rank_test_%s' % scorer] == 1)[0][0]
    best_score = -results['mean_test_%s' % scorer][best_index] if scoring[scorer]=='neg_log_loss' else results['mean_test_%s' % scorer][best_index]
        
    # Нанесём пунктирную вертикальную линию на лучший результат для этого маркера, отмеченный как x
    ax.plot([X_axis[best_index], ] * 2, [0, best_score],
            linestyle='-.', color=color, marker='x', markeredgewidth=3, ms=8)

    # Прокомментируем лучший результат для этого маркера
    ax.annotate("%0.2f" % best_score,
                (X_axis[best_index], best_score + 0.005))

plt.legend(loc="best")
plt.grid('off')
plt.show()
```

## 4.4. GridSearchCV оценивает с использованием нескольких показателей, Repeated StratifiedKFold и конвейера для предварительной обработки одновременно
Мы можем применять множество задач вместе для более углубленной оценки, например, gridsearch, используя перекрестную проверку на основе многократно повторяющихся k-повторений, которые можно масштабировать или нет по отношению ко многим показателям и настраивать параметр для данной оценки!

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.pipeline import Pipeline

#Определим простую модель
###############################################################################
C = np.arange(1e-05, 5.5, 0.1)
scoring = {'Accuracy': 'accuracy', 'AUC': 'roc_auc', 'Log_loss': 'neg_log_loss'}
log_reg = LogisticRegression()

#Простые оценки предварительной обработки
###############################################################################
std_scale = StandardScaler(with_mean=False, with_std=False)
#std_scale = StandardScaler()

#Определение метода CV: Использование повторяющегося K-кратного повторения
###############################################################################

n_folds=5
n_repeats=5

rskfold = RepeatedStratifiedKFold(n_splits=n_folds, n_repeats=n_repeats, random_state=2)

#Создание простой линии и определение поиска по сетке
###############################################################################

log_clf_pipe = Pipeline(steps=[('scale',std_scale), ('clf',log_reg)])

log_clf = GridSearchCV(estimator=log_clf_pipe, cv=rskfold,
              scoring=scoring, return_train_score=True,
              param_grid=dict(clf__C=C), refit='Accuracy')

log_clf.fit(X, y)
results = log_clf.cv_results_

print('='*20)
print("лучшие параметры: " + str(log_clf.best_estimator_))
print("лучшие параметры: " + str(log_clf.best_params_))
print('лучший результат:', log_clf.best_score_)
print('='*20)

plt.figure(figsize=(10, 10))
plt.title("GridSearchCV evaluating using multiple scorers simultaneously",fontsize=16)

plt.xlabel("Inverse of regularization strength: C")
plt.ylabel("Score")
plt.grid()

ax = plt.axes()
ax.set_xlim(0, C.max()) 
ax.set_ylim(0.35, 0.95)

# Получить обычный массив numpy из MaskedArray
X_axis = np.array(results['param_clf__C'].data, dtype=float)

for scorer, color in zip(list(scoring.keys()), ['g', 'k', 'b']): 
    for sample, style in (('train', '--'), ('test', '-')):
        sample_score_mean = -results['mean_%s_%s' % (sample, scorer)] if scoring[scorer]=='neg_log_loss' else results['mean_%s_%s' % (sample, scorer)]
        sample_score_std = results['std_%s_%s' % (sample, scorer)]
        ax.fill_between(X_axis, sample_score_mean - sample_score_std,
                        sample_score_mean + sample_score_std,
                        alpha=0.1 if sample == 'test' else 0, color=color)
        ax.plot(X_axis, sample_score_mean, style, color=color,
                alpha=1 if sample == 'test' else 0.7,
                label="%s (%s)" % (scorer, sample))

    best_index = np.nonzero(results['rank_test_%s' % scorer] == 1)[0][0]
    best_score = -results['mean_test_%s' % scorer][best_index] if scoring[scorer]=='neg_log_loss' else results['mean_test_%s' % scorer][best_index]
        
     # Нанесём пунктирную вертикальную линию на лучший результат для этого маркера, отмеченный как x
    ax.plot([X_axis[best_index], ] * 2, [0, best_score],
            linestyle='-.', color=color, marker='x', markeredgewidth=3, ms=8)

     # Прокомментируем лучший результат для этого маркера
    ax.annotate("%0.2f" % best_score,
                (X_axis[best_index], best_score + 0.005))

plt.legend(loc="best")
plt.grid('off')
plt.show()
```
