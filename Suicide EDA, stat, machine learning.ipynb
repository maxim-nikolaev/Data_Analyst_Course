#### Столбцы данных
- country: страна происхождения самоубийцы (страна, в которой произошло самоубийство).
- year: год совершения самоубийства.
- sex: пол самоубийцы.
- age: возрастная группа самоубийцы.
- suicides_no: количество совершенных самоубийств.
- population: общая численность населения в стране.
- suicides/100k pop: число самоубийств на сто тысяч человек населения.
- country-year: комбинация колонок страна и год совершения самоубийства.
- HDI for year: годовой уровень индекса человеческого развития (интегральный показатель, рассчитываемый ежегодно для межстранового сравнения и измерения уровня жизни, грамотности, образованности и долголетия как основных характеристик человеческого потенциала исследуемой территории.
- gdp_for_year ($): годовой уровень ВВП страны.

- gdp_per_capita ($): ВВП страны на душу населения.

- generation: название соответствующего поколения.

```python
df1 = df.groupby("country")["suicides_no"].sum()
country_name = list(df1.index.get_level_values(0))
len(country_name)
```

Уровень самоубийств в разбивке по странам и населению.

```python
suicide_mean_by_country=pd.DataFrame(df1.groupby(['country'])['suicides_pop'].mean())
suicide_mean_by_country = suicide_mean_by_country.sort_values(by='suicides_pop', ascending=False)
suicide_mean_by_country
```

Ранжирование по средним показателям самоубийств.

```python
suicide_median_by_country=pd.DataFrame(df1.groupby(['country'])['suicides_pop'].quantile(0.5))
suicide_median_by_country = suicide_median_by_country.sort_values(by='suicides_pop', ascending=False)
suicide_median_by_country
```

Мы можем объявить Литву самой суицидальной страной исходя из данных о самоубийствах на сто тысяч человек населения, но лучше количественно оценить результаты с некоторой статистической значимостью, прежде чем принимать решение, что мы проделаем в 4 и 5 разделах данной работы.

```python
country_list = list(df1['country'].unique())
country_suicide_ratio = []
for i in country_list:
    x = df1[df1["country"] == i]
    country_suicide_rate = sum(x.suicides_pop)/len(x)
    country_suicide_ratio.append(country_suicide_rate)
country_suicide_ratio

suicides = pd.DataFrame({"country": country_list, "country_suicide_ratio": country_suicide_ratio})
suicides=suicides.sort_values(by="country_suicide_ratio", ascending=False)
```

```python
plt.figure(figsize=(8,32))
sns.barplot(x=suicides['country_suicide_ratio'], y=suicides['country'], palette="BuPu")
plt.xlabel("Ratio of suicide")
plt.ylabel("Country")
plt.title("Suicide rate vs country")
plt.show()
```

Проверяем уровень самоубийств на страну в год и ВВП на душу населения.

```python
sorted_countries = suicides.country.tolist()
plt.figure(figsize = (10,6))
df_sui = df1.groupby(["country","year"]).suicides_pop.mean()
for country in sorted_countries[:10]:
    plt.plot(df_sui[country].index,df_sui[country].values, label=country, marker="*")
plt.xlabel("year")
plt.ylabel("suicides_pop")
plt.title("Suicide rate vs country per year")
plt.legend()
plt.show()

plt.figure(figsize = (10,6))
df_gdp = df1.groupby(["country","year"]).gdp_per_capita.mean()
for country in sorted_countries[:10]:
    plt.plot(df_gdp[country].index,df_gdp[country].values, label=country, marker="*")
plt.xlabel("Year")
plt.ylabel("Gdp_per_capita")
plt.title("GDP vs country per year")
plt.legend()
plt.show()
```

Литва является примером довольно высокой корреляции между ВВП и уровнем самоубийств.

Шри-Ланка демонстрирует снижение уровня самоубийств и почти отсутствие изменений в ВВП на душу населения. 

Российская Федерация, Венгрия, Беларусь, Латвия, Казахстан, Словения и Эстония имеют высокую корреляцию между ВВП и уровнем самоубийств - с ростом ВВП уровень самоубийств снижается. Если мы внимательно посмотрим на них, то сможем признать, что все они, за исключением Венгрии и Словении являются участниками бывшего Советского Союза, и рост числа самоубийств после краха системы был ожидаемым. # 5. Проверка гипотез.

ПРИМЕНЕНИЕ ИСПЫТАНИЙ 

ГИПОТЕЗА ДЛЯ НАИМЕНОВАНИЯ САМОЙ САМОЙ СУИЦИДАЛЬНОЙ СТРАНЫ В МИРЕ C ПОМОЩЬЮ ANOVA TEST

Мы проводим проверку гипотез, потому что при ранжировании по среднему значению Литва стоит на первом месте с гораздо более высокой ценностью, но по среднему показателю Шри-Ланка с очень небольшим отрывом опережает Литву.

Если проверка гипотез доказывает, что нет существенной разницы между показателями самоубийств в Литве и Шри-Ланке, то мы, несомненно, можем считать Литву самой суицидальной страной в мире (на основе данных за 1985–2016 годы).

НУЛЕВАЯ ГИПОТЕЗА (H0)- Между показателями самоубийств в Литве и Шри-Ланке нет существенной разницы.
АЛЬТЕРНАТИВНАЯ ГИПОТЕЗА (Н1) - Существует разница между показателями самоубийств в Литве и Шри-Ланке.

ТЕСТ СТАТИСТИКА - Разница медиан показателей самоубийств в Литве и Шри-Ланке. 
Мы принимаем размер образца 100 точек данных из каждой страны и устанавливаем уровень значимости 5% (alpha = 0,05).

ВЫПОЛНЯЕМ НАШ ЭКСПЕРИМЕНТ. 

Сначала делим данные по странам.

```python
lithuania=df1[df1.country=='Lithuania']
srilanka=df1[df1.country=='Sri Lanka']
```

Затем мы берем случайные выборки из 100 точек данных в каждой из Литвы и Шри-Ланки.

```python
sample_lithuania=lithuania.sample(100)
sample_srilanka=srilanka.sample(100)
```

Теперь мы возьмем только столбец показателей самоубийств из фрейма данных Pandas образца и преобразуем его в массив Numpy для простоты вычислений.

```python
### Массив образцов самоубийств в Литве.
S1=np.array(sample_lithuania['suicides_pop'])
print(S1)
median1=np.median(S1)
print('MEDIAN1=',median1)
```

```python
### Массив показателей самоубийств на Шри-Ланке.
S2=np.array(sample_srilanka['suicides_pop'])
print(S2)
median2=np.median(S2)
print('MEDIAN2=',median2)
```

Рассчитаем значение нашей тестовой статистики.

```python
TS= median1-median2
print("Test Statistic is (Median of Lithuania - Median of Sri Lanka)=", TS)
```

Следующим шагом является объединение двух наборов данных выборки в один набор и перетасовка их так, чтобы наблюдения обоих наборов правильно смешивались в один набор.

```python
S= np.concatenate((S1,S2), axis=None)
np.random.shuffle(S)
print(S)
```

Проводим повторную выборку 1000 раз, беря две выборки по 100 точек и каждый раз вычисляя разницу их медиан. Список медианных различий хранится в другом массиве.

```python
medianlist=np.zeros(1000)
for i in range(1000):
    _s1= np.random.choice(S,100)
    _s2= np.setdiff1d(S, _s1)
    _median1= np.median(_s1)
    _median2= np.median(_s2)
    medianlist[i]= np.array(_median1 - _median2)
```

На последнем этапе медианный список сортируется в порядке возрастания, вычисляется значение p, и мы делаем окончательный вывод.

```python
medianlist=np.sort(medianlist)
ct=0
for i in medianlist:
    if i < TS:
        ct+=1
pvalue= (1000-ct) / 1000
medianlist=pd.DataFrame(medianlist, columns=['mediandiff'])
pv95=medianlist.mediandiff[949]
print('ЗНАЧЕНИЕ В ПОЗИЦИИ 95-ГО ПРОЦЕНТИЛЯ=', pv95)
print('ТЕСТОВАЯ СТАТИСТИКА =',TS)
print('Р-ЗНАЧЕНИЕ ТЕСТОВОЙ СТАТИСТИКИ =', pvalue)
if pvalue>0.05:
    print('ЗНАЧЕНИЕ НАБЛЮДЕНИЯ НИЖЕ 95-ГО ПРОЦЕНТИЛЯ.\nТАКИМ ОБРАЗОМ ВЕРОЯТНОСТЬ ПОДТВЕРЖДЕНИЯ НАШЕГО НАБЛЮДЕНИЯ С УЧЕТОМ НУЛЕВОЙ ГИПОТЕЗЫ ВЫСОКА, ТАК КАК P-ЗНАЧЕНИЕ {} ВЫШЕ УРОВНЯ ЗНАЧИМОСТИ 0,05.'.format(pvalue))
    print('ИТАК, МЫ ПРИНИМАЕМ НУЛЕВУЮ ГИПОТЕЗУ КАК ИСТИННУЮ: В ЛИТВЕ И ШРИ-ЛАНКЕ НЕТ СУЩЕСТВЕННОЙ РАЗНИЦЫ В ПОКАЗАТЕЛЯХ САМОУБИЙСТВ.')
else:
    print('ЗНАЧЕНИЕ НАБЛЮДЕНИЯ ВЫШЕ 95-ГО ПРОЦЕНТИЛЯ.\nТАКИМ ОБРАЗОМ ВЕРОЯТНОСТЬ ПОДТВЕРЖДЕНИЯ НАШЕГО НАБЛЮДЕНИЯ С УЧЕТОМ НУЛЕВОЙ ГИПОТЕЗЫ КРАЙНЕ МАЛА, ТАК КАК P-ЗНАЧЕНИЕ {} НИЖЕ УРОВНЯ ЗНАЧИМОСТИ 0,05.'.format(pvalue))
    print('ИТАК, МЫ ОТВЕРГАЕМ НУЛЕВУЮ ГИПОТЕЗУ И ПРИНИМАЕМ АЛЬТЕРНАТИВНУЮ ГИПОТЕЗУ: ЛИТВА И ШРИ-ЛАНКА ИМЕЮТ ЗНАЧИТЕЛЬНУЮ РАЗНИЦУ В ПОКАЗАТЕЛЯХ САМОУБИЙСТВ.')
    if TS > 0:
        print('ИЗ ЗНАКА TS МЫ ЗНАЕМ, ЧТО В ЛИТВЕ БОЛЕЕ ВЫСОКИЙ УРОВЕНЬ САМОУБИЙСТВ.')
    else:
        print('ИЗ ЗНАКА ТС МЫ ЗНАЕМ, ЧТО В ШРИ-ЛАНКЕ БОЛЕЕ ВЫСОКИЙ УРОВЕНЬ САМОУБИЙСТВ.')
```

Значение p, равное 0,763, означает, что существует вероятность 76% того, что наши выборочные наблюдения будут проводиться, учитывая, что нет существенной разницы между показателями самоубийств в Литве и Шри-Ланке, что намного выше, чем уровень значимости 5%. Таким образом, мы принимаем нашу нулевую гипотезу как Истину.

Таким образом, мы можем сказать, что Литва - самая суицидальная страна в мире потому, что средняя разница в уровне самоубийств в Литве и Шри-Ланке высока, но медианная разница не значительна. Таким образом, мы выносим наше суждение, основанное на средней разнице.

Новые различные проблемы (например, Интернет и социальные сети), с которыми сталкиваются люди в некоторых странах (и которые приводят к самоубийству), компенсируют снижение уровня самоубийств в других странах, которые уже справились со своими собственными проблемами.
